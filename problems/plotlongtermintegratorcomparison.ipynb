```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(12,8))
colors = ["red", "green", "blue", "cyan"]
Ncopies = 7
for ID in range(Ncopies-1):
    for i, integ in enumerate(["implicit_midpoint", "euler", "naive", "none"]):
        try:
            data = np.loadtxt("longtermdata/"+integ+str(ID)+".txt", unpack=True)
            ax.plot(data[0]/2/np.pi, data[1], '.', color=colors[i], alpha=0.1)
        except:
            pass

for i, integ in enumerate(["implicit_midpoint", "euler", "naive", "none"]):
        try:
            data = np.loadtxt("longtermdata/"+integ+str(ID)+".txt", unpack=True)
            ax.plot(data[0]/2/np.pi, data[1], '.', color=colors[i], label=integ, alpha=0.1)
        except:
            pass
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim([1e-10,0.1])
ax.legend(loc='lower left')
ax.set_xlabel('Time (orbits)')
ax.set_ylabel('Relative Hamiltonian Error')
```
