```python
import rebound
import reboundx
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import re
```

```python
import os
K, mag, t = [], [], []
for filename in os.listdir('data/'):
    result = re.search('IC(.*)K(.*)mag(.*).bin', filename)
    if result:
        sim = rebound.Simulation.from_archive('data/'+filename)
        K.append(float(result.group(2)))
        mag.append(float(result.group(3)))
        t.append(sim.t)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
cm = plt.cm.get_cmap('RdYlBu')
sc = ax.scatter(K, mag, c=np.log10(t), edgecolor='', vmin=3, s=80,cmap=cm)        
ax.set_xlabel('K', fontsize=24)
ax.set_ylabel('mag', fontsize=24)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim([10,1000])
ax.set_ylim([1.e-3, 1])
plt.colorbar(sc)
```
